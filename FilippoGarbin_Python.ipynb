```python
vetture=[
    {'nome':'Jeep Avenger','ratings':{'score':97,'clean_air':10,'energy_efficiency':9.4,'greenhouse_gas':0},'prezzo':38450,'motore':'elettrico'},
    {'nome':'Honda e:Ny1','ratings':{'score':94,'clean_air':10,'energy_efficiency':9,'greenhouse_gas':0},'prezzo':54040,'motore':'elettrico'},
    {'nome':'Honda Civic','ratings':{'score':56,'clean_air':7.2,'energy_efficiency':5.5,'greenhouse_gas':113},'prezzo':39240,'motore':'ibrido'},
    {'nome':'Mercedes Benz GLC','ratings':{'score':55,'clean_air':9.4,'energy_efficiency':4.5,'greenhouse_gas':144},'prezzo':65319,'motore':'diesel'},
    {'nome':'Skoda Karoq','ratings':{'score':52,'clean_air':7.9,'energy_efficiency':4.7,'greenhouse_gas':135},'prezzo':24808,'motore':'benzina'},
    {'nome':'MINI Countryman','ratings':{'score':46,'clean_air':7.9,'energy_efficiency':3.9,'greenhouse_gas':153},'prezzo':40900,'motore':'benzina'},
    {'nome':'Subaru Crosstrek','ratings':{'score':39,'clean_air':6.7,'energy_efficiency':3.3,'greenhouse_gas':174},'prezzo':34254,'motore':'ibrido'},
    {'nome':'Toyota bZ4X','ratings':{'score':97,'clean_air':10,'energy_efficiency':9.4,'greenhouse_gas':0},'prezzo':40150,'motore':'elettrico'},
    {'nome':'Kia EV6','ratings':{'score':94,'clean_air':10,'energy_efficiency':9,'greenhouse_gas':0},'prezzo':48000,'motore':'elettrico'},
    {'nome':'BYD SEAL U','ratings':{'score':93,'clean_air':10,'energy_efficiency':8.9,'greenhouse_gas':0},'prezzo':38900,'motore':'elettrico'},
    {'nome':'Peugeot 308','ratings':{'score':55,'clean_air':6.5,'energy_efficiency':5.6,'greenhouse_gas':119},'prezzo':32070,'motore':'diesel'},
    {'nome':'VW Taigo','ratings':{'score':53,'clean_air':7.0,'energy_efficiency':5.2,'greenhouse_gas':135},'prezzo':24808,'motore':'benzina'},
    {'nome':'Kia Ceed','ratings':{'score':37,'clean_air':4.4,'energy_efficiency':4.2,'greenhouse_gas':135},'prezzo':24808,'motore':'benzina'},
    {'nome':'Honda CRV','ratings':{'score':44,'clean_energy':7.3,'energy_efficiency':3.7,'greenhouse_gas':151},'prezzo':49240,'motore':'ibrido'},
    {'nome':'FIAT Doblo','ratings':{'score':42,'clean_energy':8.6,'energy_efficiency':2.8,'greenhouse_gas':166},'prezzo':21700,'motore':'diesel'}
]
    
def stampa_listino():                                                                                  #Funzione per la visualizzazione di tutti gli elementi.
    while True:                                                                                        #Inizia un ciclo while.
        x = input('Desideri vedere tutto il listino? Rispondi con Y/N:').lower()                       #Chiede una risposta all'utente e lo abbina ad una variabile.
        if x =='y':                                                                                    #Se la risposta è positiva, stampa tutti i dizionari della lista come se fossero liste; per essere più leggibile.
            for dict in vetture:
                print(list(dict.items()))
            break                                                                                      #Interrompe il ciclo while dopo aver stampato tutti i dizionari della lista vetture.
        elif x =='n':
            print('Operazione terminata')
            break                                                                                      #Se la risposta è negativa, stampa un messaggio solo e interrompe il ciclo while. 
        else: 
            print('Risposta non valida.')
            continue                                                                                   #Se la risposta è un valore inatteso, stampa un messaggio e continua il ciclo, finchè una delle prime due condiioni non si avvera.  
stampa_listino()                                                                                      #Richiama la funzione.

def nuova_vettura(nome, score, clean_air, energy_efficiency, greenhouse_gas, prezzo, motore):          #Funzione per inserire una nuova vettura alla lista vetture. Vengono esplicitati gli attributi della funzione.
    ratings={
        'score':score,
        'clean_air':clean_air,
        'energy_efficiency':energy_efficiency,
        'greenhouse_gas':greenhouse_gas
    }                                                                                                  #Vengono dichiarate le coppie chiave-valore del dizionario annnidato 'ratings'.
    vetture.append({
        'nome':nome,
        'ratings':ratings,
        'prezzo':prezzo,
        'motore':motore
    })                                                                                                #Vengono aggiunte tutte le coppie chiave-valore (compresa 'ratings') alla lista vetture. 
    print('Ecco l\'ultima auto aggiunta!',vetture[-1])                                                #Stampa un messaggio e l'ultimo dizionario della lista vetture che abbiamo appena aggiunto.
    

nuova_vettura('Smart #3',95,10,9.2,0,38548,'elettrico')                                             #Richiama la funzione insieme agli attributi della nuova vettura.

def cerca_punteggio():                                                                               #Funzione per cercare il valore di valutazione generale ('score') dell'auto rilasciato da Green NCAP.
    while True:                                                                                      #Inizia un ciclo while.
        try:                                                                                         #'Try' permette alla funzione di accettare valori diversi rispetto a un numero.
            s = int(input('Cerca un punteggio Green NCAP da 0 a 100:'))                              #Chiede un valore all'utente e lo abbina ad una variabile
            if 0<= s <=100:
                risposta = False                                                                     #La variabile 'risposta' serve nel caso il valore inserito non sia presente.
                for dict in vetture:                                                                 #Ciclo for annidato
                  if dict['ratings']['score']==s:                                                    #Abbina la risposta dell'utente ('s') ai valori 'score' del dizionario.
                      print(f'L\'automobile {dict['nome']} ha ottenuto un punteggio Green NCAP pari a: {s}') #Se il valore inserito dall'utente è presente, stampa tutti i dizionari in cui è presente.
                      risposta = True                                                                #La variabile 'risposta' diventa vera e il ciclo for termina.
                if not risposta:                                                                     #Se il valore inserito dall'utente non è presente, 'risposta' rimane falsa e l'utente viene notificato con un messaggio.
                    print(f'Nessuna auto ha ottenuto il punteggio Green NCAP pari a {s}')
                break                                                                                #Break quindi interrompe il ciclo for annidato.
        except ValueError:                                                                           #'Except', insieme a 'try' serve se l'utente inserisce un valore diverso dal tipo 'int'.
            print('Risposta non valida. Inserisci un numero compreso tra 0 e 100')                   #Invece di segnalare un errore e bloccare la funzione, viene mostrato un messaggio, e il ciclo while continua.
        

cerca_punteggio()                                                                                   #Richiama la funzione.

def statistiche():                                                                                   #Funzione per calcolare quante auto sono presenti nella lista per tipologia di motorizzazione.
    conta_elettrico = conta_ibrido = conta_benzina = conta_diesel = 0                                #Crea 4 contatori diversi e li inizalizza a 0.
    for dict in vetture:                                                                             #Ciclo for per creare gli abbinamenti tra i contatori e le chiavi 'motore'.
        tipo_motore = dict['motore']
        if tipo_motore=='elettrico':
             conta_elettrico+=1                                                                      #Per ogni 'if', a seconda della corrispondenza, ogni contatore viene aumentato di 1. Questo avviene per tutte le vetture.
        if tipo_motore=='ibrido':
             conta_ibrido+=1
        if tipo_motore=='benzina':
             conta_benzina+=1
        if tipo_motore=='diesel':
             conta_diesel+=1 
    
    print(f'Sono presenti {conta_elettrico} vetture elettriche, {conta_ibrido} vetture ibride, {conta_benzina} vetture a benzina e {conta_diesel} vetture a diesel.')    #Stampa un messaggio unico dove mostra il totale di auto per tipologia di motore.

statistiche()                                                                                       #Richiama la funzione.
        
def filtraggio_avanzato():                                                                           #Funzione per filtrare tutte le auto che rispetano una condizione doppia.
    for dict in vetture:                                                                             #Ciclo for per iterare tutti i dizionari.
        costo = dict['prezzo']                                                                       #Crea due variabili da abbinare alle condizioni che voglio cercare: 'prezzo' e 'greenhouse_gas'.
        anidride = dict['ratings']['greenhouse_gas']
        if 24000 < costo < 25000 and 130 < anidride < 140:                                           #Imposta la doppia condizione con 'if' e 'and'.  
            print(f'La {dict['nome']} ha un prezzo pari a {costo} euro e produce {anidride} grammi di CO2 al km.')   #Stampa un messaggio.
            
filtraggio_avanzato()                                                                               #Richiama la funzione.











        





          
                               
    
```
